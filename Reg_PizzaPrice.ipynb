**Splitting the Data**

```python
X = df.drop(columns=['price_rupiah', 'extra_sauce', 'extra_cheese', 'extra_mushrooms', 'size', 'variant', 'topping', 'company'])
y = df['price_rupiah']
```

```python
X
```

```python
X.columns
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_scaled = sc_x.fit_transform(X)
```

```python
sc_y = StandardScaler()
y_scaled = sc_y.fit_transform(y.values.reshape(-1, 1))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)
```

# Building the Model

```python

from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error
from sklearn.ensemble import  RandomForestRegressor
from xgboost import XGBRFRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.pipeline import Pipeline
```

```python
lr = LinearRegression()
svr = SVR()
rfr = RandomForestRegressor()

xgb = XGBRFRegressor()
```

```python
param1 = {}
param1['regressor'] = [lr]

param2 = {}
param2['regressor__C'] = [10**-2, 10**-1, 10**0, 10**1, 10**2]
param2['regressor'] = [svr]

param3 = {}
param3['regressor__n_estimators'] = [50, 100, 200, 500]
param3['regressor__max_depth'] = [5, 10, 20]
param3['regressor'] = [rfr]

param4 = {}
param4['regressor'] = [xgb]
param4['regressor__max_depth'] = [5, 8,10, 20]
param4['regressor__n_estimators'] = [10,50, 100, 200, 500]

```

```python
pipeline = Pipeline([('regressor', lr)])
params = [param1, param2, param3, param4]
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(pipeline, params, cv=5, scoring='r2')

```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

# Model Evaluation


```python
from sklearn import metrics
import math
```

```python
y_pred = my_model.predict(X_test)
y_pred_orig = sc_y.inverse_transform(y_pred.reshape(-1, 1))
y_pred_orig
```

```python
test_r2_score = metrics.r2_score(y_test, y_pred) # R2_score
print(f'Test R2_score: {test_r2_score}')

test_mse = metrics.mean_squared_error(y_test, y_pred) # Mean Squared Error Score
print(f'Test MSE : {test_mse}')

test_RMSE = math.sqrt(metrics.mean_squared_error(y_test, y_pred)) # SQRT MSE Score
print(f'Test RMSE : {test_RMSE}')
```
